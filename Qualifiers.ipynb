```python
q1_time_df = sorted_qualifier_columns['q1_time']
q2_time_df = sorted_qualifier_columns['q2_time']
q3_time_df = sorted_qualifier_columns['q3_time']

sorted_qualifier_columns.drop(['q1_time', 'q2_time', 'q3_time', 'qualifyId'], axis=1, inplace=True)
```

```python
q1_time_ms = []
q2_time_ms = []
q3_time_ms = []

for i in q1_time_df:
    if(i != '\\N' and i==i):
        x = datetime.strptime(i,'%M:%S.%f')
        q1_time_ms.append((x.minute*60*100) + (x.second*100) + (x.microsecond/1000))
    else:
        q1_time_ms.append(0)

for i in q2_time_df:
    if(i != '\\N' and i==i):
        x = datetime.strptime(i,'%M:%S.%f')
        q2_time_ms.append((x.minute*60*100) + (x.second*100) + (x.microsecond/1000))
    else:
        q2_time_ms.append(0)

for i in q3_time_df:
    if(i != '\\N' and i==i):
        x = datetime.strptime(i,'%M:%S.%f')
        q3_time_ms.append((x.minute*60*100) + (x.second*100) + (x.microsecond/1000))
    else:
        q3_time_ms.append(0)

q1_time_ms = pd.Series(q1_time_ms)
q2_time_ms = pd.Series(q2_time_ms)
q3_time_ms = pd.Series(q3_time_ms)
```

```python
q1_time_max = q1_time_ms.max()
q2_time_max = q2_time_ms.max()
q3_time_max = q3_time_ms.max()

q1_time_ms = q1_time_ms.replace(0, q1_time_max*2)
q2_time_ms = q2_time_ms.replace(0, q2_time_max*2)
q3_time_ms = q3_time_ms.replace(0, q3_time_max*2)

temp_qualifier1 = pd.concat((sorted_qualifier_columns.reset_index(drop=True), q1_time_ms.rename('q1_time_ms').reset_index(drop=True)), axis=1)
temp_qualifier2 = pd.concat((temp_qualifier1, q2_time_ms.rename('q2_time_ms')), axis=1)
temp_qualifier3 = pd.concat((temp_qualifier2, q3_time_ms.rename('q3_time_ms')), axis=1)
temp_qualifier3
```

```python
temp_qualifier3.to_csv('./qualifier_data/qualify.csv')
```
