```python
import pandas as pd
import numpy as np
```

```python
dataset = pd.read_csv('./data/dataset.csv')
dataset.columns
```

```python
dataset.shape[0]
```

```python
laps_columns = dataset[['raceId', 'year', 'circuitId', 'driverId', 'constructorId', 'laps', 'fastestLap', 'fastestLapTime', 'fastestLapSpeed', 'first_5_avg_time', 'last_10_avg_time', 'lap_position_after_5', 'lap_position_before_last_10']]
laps_columns
```

```python
pit_stops_columns = dataset[['raceId', 'year', 'circuitId', 'driverId', 'constructorId', 'no_of_stops', 'ms_avg']]
pit_stops_columns
```

```python
new_laps_columns = laps_columns.dropna(axis=0, subset=['last_10_avg_time'])
sorted_laps_columns = new_laps_columns.sort_values(by=['raceId'], ascending=True)
sorted_laps_columns
```

```python
new_pit_stops_columns = pit_stops_columns.dropna(axis=0, subset=['ms_avg'])
sorted_pit_stops_columns = new_pit_stops_columns.sort_values(by=['raceId'], ascending=True)
sorted_pit_stops_columns
```

```python
laps_columns_no_n = sorted_laps_columns.loc[sorted_laps_columns['fastestLap'] != '\\N']
laps_columns_no_n
```
